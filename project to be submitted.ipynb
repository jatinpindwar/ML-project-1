### Print unique values of each column of train and test data 


```python
#EXTRACT FEATURES FROM TRAIN DATA
features=train_data.drop(['ID'],axis=1)
```

```python
features.head()
```

```python
#UNIQUE VALUES FOR EACH FEATURES AND TEST DATA AND STORE THEM IN ARRAY
train_unique=[]
train_unique_count=[]
for i in features.columns:
    print('NAME OF COLUMN IS : ' ,i,'--HAVING UNIQUE VALUES--',features[i].unique(),'WITH COUNT--',len(features[i].unique()))
    print('\n')
    train_unique.append(features[i].unique())
    train_unique_count.append(len(features[i].unique()))
    
```

```python
test_data1=test_data.drop('ID',axis=1)
```

```python
#FOR TEST DATA
test_unique=[]
test_unique_count=[]
for i in test_data1.columns:
    print('NAME OF COLUMN IS : ' ,i,'--HAVING UNIQUE VALUES--',test_data1[i].unique(),'WITH COUNT--',len(test_data1[i].unique()))
    test_unique.append(test_data1[i].unique())
    test_unique_count.append(len(test_data1[i].unique()))
    print('\n')
    
```

### Make Dataframe of unique values and their count

```python
train_values_df=pd.concat([pd.DataFrame(features.columns),pd.Series(train_unique),pd.Series(train_unique_count)],axis=1)
train_values_df.columns=train_values_df.columns=['column_name','unique_value_array','count']
test_values_df=pd.concat([pd.DataFrame(test_data1.columns),pd.Series(test_unique),pd.Series(test_unique_count)],axis=1)
test_values_df.columns=['column_name','unique_value_array','count']
```

### Information retrieve from unique values count of train and test data----

WE CAN SEE THAT SOME OF COLUMNS OF TEST DATA HAVE UNIQUE VALUES COUNT GREATER THAN THAT OF TRAIN DATA COLUMNS.

### Check for zero variance columns in train  data set and remove those columns from both train and test dataset

```python
#CHECK FOR ZERO VARIANCE IN FEATURES AND DROP THOSE COLUMNS FROM BOTH DATA SET
#FEATURES
col_zerovar=train_data.var()==0.0
a=col_zerovar[col_zerovar==True]
a.index
```

```python
new_features=train_data.drop(a.index,axis=1)
```

```python
new_features.shape
```

```python
#DROP SAME COLUMNS FROM TEST DATASET AS WE DONT BUILD OUR MODEL ON THOSE ZERO VARIANCE FEATURES
test_data1=test_data.drop(a.index,axis=1)
```

```python
test_data1.shape
```

### Outlier detection and removal in train dataset 

```python
##OUTLIER DETECTION AND REMOVAL FOR TRAIN DATA ON BASIS OF Y LABEL
sns.boxplot(data=train_data.y)
```

```python
train_data.describe()
```

```python
## Y - TO FIND UPPER BOUND VALUE OF BOXPLOT
Q1=90.82
Q3=109.01
Q2=99.15
IQR=Q3-Q1
upper_bound=Q3+1.5*IQR
print(upper_bound)
```

```python
## REMOVE ZERO VARIANCE ALREADY AND NOW REMOVE OUTLIERS
train_no_outlier=new_features[new_features.y<upper_bound]
train_no_outlier.shape
```

### Label Encoding of Categorical Columns

```python
## SEPERATE OBJECT TYPE COLUMNS OF BOTH DATASET FOR LABEL ENCODING
features_object=train_no_outlier[train_no_outlier.columns[train_no_outlier.dtypes=='object']]
```

```python
features_object.shape
```

```python
features_object.head()
```

```python
##FOR TEST DATA OBJECT TYPE COLUMNS EXTRACTION
test_object=test_data1[test_data1.columns[test_data1.dtypes=='object']]
test_object.shape
```

```python
test_object.head()
```

```python
#ENCODE THESE LABELS USING LABEL ENCODER
label=LabelEncoder()
```

```python
# TRAIN OBJECT FEATURES TRANSFORMATION
for i in features_object.columns:
    features_object[i]=label.fit_transform(features_object[i])
    test_object[i]=test_object[i].map(lambda x: 'misc' if x not in label.classes_ else x)
    label.classes_=np.append(label.classes_,'misc')
    test_object[i]=label.transform(test_object[i])
```

### Extract remaining data for dimensionality reduction

```python
#CATEGORICAL FEATURES EXTRACTION OF TRAIN AND TEST DATA
features_cat=train_no_outlier[train_no_outlier.columns[train_no_outlier.dtypes!='object']]
test_cat=test_data1[test_data1.columns[test_data1.dtypes!='object']]
```

```python
features_cat.shape,test_cat.shape
```

```python
#DROP Y COLUMN FROM features_cat
features_cat.drop('y',axis=1,inplace=True)
```

```python
features_cat.set_index('ID',inplace=True)
features_cat.head()
```

```python
test_cat.set_index('ID',inplace=True)
test_cat.head()
```

### DIMENSIONALITY REDUCTION USING PCA

#### As we know that pca not use for categorical data dimension reduction but as we said in the problem to perform dimensionality reduction that's why we perform PCA.

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.995)
```

```python
red_train=pd.DataFrame(pca.fit_transform(features_cat))
pca.explained_variance_ratio_
```

```python
print(red_train.shape)
print(features_object.shape)
print(train_no_outlier.shape)
```

```python
red_train.index,features_object.index
```

```python
red_train.reset_index(inplace=True)
features_object.reset_index(inplace=True)
```

```python
final_data=pd.merge(features_object,red_train,on='index')
final_data.shape
```

```python
final_data.head()
```

```python
red_test=pd.DataFrame(pca.transform(test_cat))
```

```python
print(red_test.shape)
print(test_object.shape)
```

```python
#NOW COMBINE OBJECT REDUCE TEST DATASET
red_test['ID']=test_data['ID']
test_object['ID']=test_data['ID']
test_predict=pd.merge(test_object,red_test)
```

```python
test_predict.shape
```

```python
test_predict.head()
```
